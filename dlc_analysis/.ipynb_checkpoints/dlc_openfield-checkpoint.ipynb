```python
fig = plt.plot(velocity_df['centroid'])
ax = plt.gca()
def frames_to_time(x):
    return(x*0.05)
def time_to_frames(x):
    return(x*20)
secax = ax.secondary_xaxis('top', functions=(frames_to_time, time_to_frames))
plt.show()
```
